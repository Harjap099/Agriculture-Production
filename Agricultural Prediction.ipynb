Now, we apply Gradient Decent using one Independent and one Dependent Variable.

```python
x_data= df.dropna().loc[: ,['Crop_Year']].values
sample= len(x_data)
x_ones= np.ones((sample, 1))
x_final= np.append(x_ones , x_data, axis =1)
y_final = df.dropna().loc[: ,['Production']].values
```

In this, case we need Y_Final as 'Production' as we need to predict it w.r.t 'Crop_Year'.

```python
new_beta=np.random.rand(2,1)
```

```python
for i in range(13):
    y_pred= np.dot(x_final , new_beta)
    derivative = -2*(np.dot(np.transpose(y_final- y_pred),x_final))/sample
    new_beta = new_beta-(0.1*derivative.T)
    cost= ((y_final-y_pred)**2)/sample                     
```
